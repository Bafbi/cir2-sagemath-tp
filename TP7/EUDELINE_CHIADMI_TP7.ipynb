1) Si on a les suites binaire de taille n, pour avoir celle de taille n+1, il suffit de rajouter 0 a chacune de nos suite de taille n, puis d'ajouter nos suites avec 1 en plus.

````python
def suite ( n : int ) :
    # condition d'arret
    if n == 1:
        return [(0,),(1,)]
    # appel recursif
    else :
        # recuperer la suite n-1
        previous_suite = suite (n -1)
        # construire la suite n
        new_suite = []
        for tuple in previous_suite :
            new_suite.append ( tuple + (0,) )
            new_suite.append ( tuple + (1,) )
        return new_suite
````

````python
lenght = 5
print(f"Pour une taille de {lenght}, on obtient une suite de {len(suite(5))} tuples binaires : {suite(5)}")
````

1) On prend nos parties d'ensemble auquelle on rajoute nos parties d'ensemble avec un 3 ajouté a chaque parties

````python
def parties ( ensemble : list ) :
    # condition d'arret
    if len (ensemble) == 1:
        return [ensemble] + [[]]
    # appel recursif
    else :
        # recuperer les sous-partitions
        previous_suite = parties (ensemble [1:])
        # print(previous_suite)
        # construire les partitions
        new_suite = []
        for subpartition in previous_suite :
            new_suite.append ( [ensemble [0]] + subpartition )
            new_suite.append ( subpartition )
        
        return new_suite
    
````

````python
ensemble = [1,2,3,4]
parties_calculees = parties(ensemble)
print(f"Pour la liste {ensemble}, on obtient les {len(parties_calculees)} parties suivantes : {parties_calculees}")
print(2**len(ensemble))
````

## Exercice 3 : parties de taille donnée

Si on reprend l’ensemble renvoyé par parties(E) dans le cas E = {1, 2, 3}, on se rend compte qu’il y a :
1. Une partie de E à 0 élément qui est ∅
2. Trois parties de E à 1 élément qui sont {1}, {2} et {3}
3. Trois parties de E à 2 éléments qui sont {1, 2}, {2, 3} et {1, 3}
4. Une partie de E à 1 élément qui est E

Dans le cas où E est un ensemble à n ∈ $\mathbb{N}$ éléments, le nombre de parties de E à k ∈ ⟦0, n⟧ éléments est
donnée par :

$$\begin{pmatrix}
n \\
k
\end{pmatrix} = \frac{n!}{k!(n - k)!}$$

<span style="color:blue">Répondez aux questions suivantes.</span><br>
<span style="color:blue">1. Générer l’ensemble parties_taille(k,E) des parties de taille k d’un ensemble E en sélectionnant
dans parties(E) les parties de taille requise. Vérifier que vous trouvez le bon nombre de parties
avec la commande binomial(n,k).</span><br>
<span style="color:blue">2. Pour construire les parties à k éléments d’un ensemble E à n éléments, on peut :<br>
— Distinguer un élément a ∈ E.<br>
— Construire les parties à k éléments de E ne contenant pas a. Cela revient à construire les parties
à k éléments de E \ {a}.<br>
— Construire les parties à k éléments de E contenant a. Cela revient à prendre les parties à k − 1
éléments de E \ {a} puis à ajouter a pour avoir les parties à k éléments.
<br>En termes de coefficient binomial, on vient de montrer :<br>
$$\begin{pmatrix}
n \\
k
\end{pmatrix} = \begin{pmatrix}
n-1 \\
k
\end{pmatrix} + \begin{pmatrix}
n-1 \\
k-1
\end{pmatrix}
$$</span>

<span style="color:blue">Utiliser cette astuce pour avoir une implémentation récursive de parties_taille(k,E).<br>3. Comparer les approches a) et b). Mettre %%time en début de cellule d’exécution pourrait vous aider.</span>

````python
def parties_taille ( ensemble : list , k : int ) :
    # On récupère toutes les sous-partitions
    subpartitions_list = parties(ensemble)
    # On filtre les sous-partitions de taille k
    return [subpartition for subpartition in subpartitions_list if len(subpartition) == k]
````

````python
length = 3
ensemble = [1,2,3,4]
parties = parties_taille(ensemble, length)
size = binomial(len(ensemble),length)

print(f"{size} parties of length {length} in {ensemble} : {parties}")
````

````python
def parties_taille3 ( ensemble : list , k : int ) :
    # condition d'arret
    # si k = 1 alors on renvoie les parties de taille 1 qui sont les éléments de E
    if k == 1:
        return [[element] for element in ensemble]
    # si k = |E| alors on renvoie la seule partie de taille |E| qui est E
    elif len (ensemble) == 1:
        return []
    # appel recursif
    else :
        # recuperer les sous-partitions de E \ {a}
        previous_suite = parties_taille3 (ensemble [1:], k)
        new_suite = []
        for subpartition in previous_suite :
            new_suite.append ( subpartition )
           
        # recuperer les sous-partitions de taille k-1 
        previous_suite = parties_taille3 (ensemble [1:], k-1)
        # print(k, previous_suite)
        for subpartition in previous_suite :
            # ajouter {a} à chaque sous-partition
            new_suite.append ( [ensemble [0]] + subpartition )
        return new_suite
````
