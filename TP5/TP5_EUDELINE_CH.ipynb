```python
def plot_serie(un, N, k) :
    S = 0.
    sommes_partielles = [ S ]
    termes = []

    for n in [k.. N]:
        S += un(n)
        sommes_partielles.append ( S )
        termes.append (un(n))

    # only plot the first 100 partial sums and terms
    img = list_plot( sommes_partielles [:N+1])
    img += plot(sommes_partielles[N] , (0 ,N) , color = 'red')
    img += list_plot( termes [:N+1] , color = 'green')
    show(img)
```
