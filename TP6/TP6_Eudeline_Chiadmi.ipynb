````python
n = var('n')

f(n) = z^n/2^n
s(z) = 1/2 + 3/4 * z + 7/2 * sum(f(n), n, 2, 100)
plotComplex(s, (-2.5, 2.5), (-2.5,2.5))
````

On a ici : $\frac 1 2 + \frac 3 4 z + \frac 7 2 \sum_{n=2}^{+∞} \frac{z^n}{2^n}$  
Graphiquement on peut determiner que sont domaine de definition est de [-2, 2] sur l'axe des réels et [-2, 2] sur l'axe des Imaginaires.  
La série entiere à donc un rayon de convergence de R=2  centré sur (0, 0). 

````python
f(n) = ((-1)^n/3^(n+1)) * (z - 3)^n


s(z) = f(0)
for i in [1..100]:
    s(z) = s(z) + f(i)
# s(z) = sum(f(n), n, 0, 100)
plotComplex(s, (-0.5, 6.5), (-3.5, 3.5))
````

On a ici : $\sum_{n=0}^{+∞} \frac{(-1)^n}{3^{n+1}}\; (z-3)^n$  
Graphiquement on peut determiner que sont domaine de definition est de [0, 6] sur l'axe des réels et [-3, 3] sur l'axe des Imaginaires.  
La série entiere à donc un rayon de convergence de R=3 centré sur (0, 3).

````python
f(n) = ((-1)^n/2*n+1) * z^(2*n + 1)
s(z) = sum(f, n, 0, 200)
plotComplex(s, (-1.5, 1.5), (-1.2,1.2))
````

On a ici : $\sum_{n=0}^{+∞} \frac{(-1)^n}{2n+1}\; z^{2n+1}$  
Graphiquement on peut determiner que sont domaine de definition est de [-1, 1] sur l'axe des réels et [-1, 1] sur l'axe des Imaginaires.  
La série entiere à donc un rayon de convergence de R=1 centré sur (0, 0).

$$z \mapsto \frac{1}{z}$$
$$\frac{1}{z} = \frac{1}{3 + (x - 3)}
= \sum \frac{(-1)^n (z-3)^n}{3^{n+1}}$$

Pour celle ci, on va partir du developpement :  
On regarde juste la série pour le moment :
$$\sum_{-2}^{+∞} \frac{z^n}{2^n}
= \sum_{-2}^{+∞} (\frac{z}{2})^n
= \frac{2}{2-z} - 1 - \frac{z}{2}$$
On revient sur le developpement complet  
$$\frac{1}{2} + \frac{3}{4}z + \frac{7}{2} \sum_{-2}^{+∞} \frac{z^n}{2^n}
=\frac{1}{2} + \frac{3}{4}z + \frac{7}{2}(\frac{2}{2-z} - 1 - \frac{z}{2})$$
$$= \frac{1}{2} + \frac{3}{4}z + \frac{7}{2-z} - \frac{7}{2} - \frac{7}{4}z
=-3 -z + \frac{7}{2-z}$$
$$=\frac{-3(2-z)}{2-z} - \frac{2z - z^2}{2-z} + \frac{7}{2-z}
=\frac{z^2 + z + 1}{2 -z}$$
    

$$z \mapsto arctan(z)$$

=\int \sum_{n=0}^{+∞} (-1)^n z^{2n}$$
$$=\sum_{n=0}^{+∞} (-1)^n \int z^{2n}
=\sum_{n=0}^{+∞} (-1)^n \frac{z^{2n+1}}{2n+1} + C$$

## Exercice 3 : avec une autre carte

De façon générale, on peut obtenir de jolies (et instructives) figures de la façon suivante.
1. On choisit une image g: $\mathbb{C} \;\to \;\chi \; où \;\chi$ est un espace de couleurs, par exemple $[0,255]^3$ si on travaille en RGB.
2. On fabrique l’image g ◦ f c’est-à-dire que chaque point z du plan est colorié avec la couleur g(f(z)).Ce qu’on obtient est une visualisation def(z).
L’image utilisée par défaut par Sage s’obtient en visualisation z $\to$ f(z).

```py
z = var('z')
f(z) = z # which  is  decomposed  into r * exp(i theta)
complex_plot(f, (-2,2), (-2,2), aspect_ratio =1)
```

On peut bien sûr choisir ce que l’on veut pour la fonction g. Par exemple, avec le quadrillage :
$$z\to g(z) \: = \: sin[\pi Re(z)]\: × \: sin[\pi Im(z)]$$

<span style="color:blue">Visualiser ce quadrillage puis utiliser le pour visualiser les fonctions précédentes. Que pouvez-vous diresur la structure générale de la visualisation ? Le mot cléconforme devrait vous aider dans vos recherches internet.</span>\:
[Applications-conformes](https://images.math.cnrs.fr/Applications-conformes.html)

````python
z = var('z')
def quad(f, im = ( -2 ,2), re = ( -2 ,2)):
    g(z) = sin(pi*real(z)) * sin(pi*imaginary(z))
    show(complex_plot(g(f), im, re, aspect_ratio =1))
````

## Bonus : l’hyptohèse de Riemann

Si vous souhaitez gagner un million de dollars, attelez-vous à la résolution d’un des sept problèmes dumillénaires :*l’hypothèse de Riemann*. Soit ζ la fonction définie par :

$$\forall\; s \;\in \;\mathbb{C}\; tel \;que \;Re(s)\; \geq 1, \;ζ(s) = \sum_{n=1}^{+∞}\frac {1} {n^s}$$
que l’on prolonge analytiquement sur tout $\mathbb{C}$. L’hypothèse de Riemann dit que tous les zéros non triviaux de ζ ont pour partie réelle \frac 1 2 . Ce résultat, observé graphiquement mais jamais démontré, permetde mieux connaître la répartition des nombres premiers. C’est quand même un truc dedingue: l’analyse et l’arithmétique qui semblent bien éloignées sont en fait liés, via la théorie analytique des nombres. <span style="color:blue"> Visualiserla fonction ζ, déjà connue sous Sage en appelant zeta. Si le compte-rendu du TP est terminé, je vous encourage à regarder l’excellente vidéo :https://www.youtube.com/watch?v=dNpdMYB8pZs.</span>
